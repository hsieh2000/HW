1.	因業務需求，客戶需要調出第61題的填答狀況，並加以分析。
該題要求受訪者勾選有使用的影音平台。
※此題為多選題
第1個選項為「沒有使用任何影音平台」，若勾選此選項，則該欄位標記為1，且其餘9個選項皆無法選取。
選項2~10分別為9個常見的影音平台，若使用者有勾選該平台，則標記為1。

Q1. 請算出每名受訪者有使用的平台總數，並儲存在新的變數欄位 B2_Count


```python
# 透過迴圈取得與問題61有關的欄位位置
question61_columns = [col for col, name in enumerate(RAW_DATA.columns) if "q0061" in name]

# 篩選出與問題61有關的欄位
question_61 = RAW_DATA.iloc[:, question61_columns[0]:question61_columns[-1]+1]
question_61
```

```python
# 計算每名受訪者有使用的平台總數，將所有空值補0，如果勾選第一個選項則"B2_Count"欄位值等於0，如果全部欄位總和為0代表受訪者未填寫此題；則"B2_Count"欄位值等於Nan，
# 其他情形則透過加總除第一個選項以外的欄位正常計算平台數，並將計算結果存入"B2_Count"欄位
question_61["B2_Count"] = question_61.fillna(0).apply(lambda x : 0 if x[0] == 1 else (np.nan if sum(x) == 0 else sum(x[1:])), axis = 1)
```

2.	第63題詢問受訪者對於有在使用的影音平台的喜好程度，評分從0分到10分。(若該受訪者有使用A,B,C三個平台，則此題僅會詢問這三個平台的喜好程度)。
客戶想標記出有「隨意填答」疑慮的受訪者，判斷標準如下：
”Type1”：所有影音平台的喜好度都相同，例如：5,5,5
“Type2”：交錯填答，例如：1,3,1,3,1,3 或 2,4,6,4,2,4 這類填答間距相同的情形
“Type3”：超過9成的喜好度填答相同，例如：1,1,5,1,1,1,1,1,1,1

Q2. 請建立一個函式，判斷受訪者填答屬於哪一種Type，並將每位受訪者的Type儲存在新的欄位Status中。
如不屬於以上任何一種Type，請標記為”PASS”，如受訪者選取的平台不超過兩個，也標記為”PASS”


```python
# 透過迴圈取得與問題63有關的欄位位置
question63_columns = [col for col, name in enumerate(RAW_DATA.columns) if "q0063" in name]

# 篩選出與問題63有關的欄位
question_63 = RAW_DATA.iloc[:, question63_columns[0]:question63_columns[-1]+1]
question_63
```

```python
# 定義Sum_dict函式，用於問題2中Type2，計算使用者填答間距是否相同
def gap(row):
  lst = [i for i in list(row) if i != -1]         #排除值為-1(透過fillna將空值替換為-1)的值
  gap_lst = [abs(lst[i+1]-lst[i]) for i in range(len(lst)-1)] #計算兩兩填答之間的絕對值間距
  return gap_lst  #回傳的值為一個list，裡面包含兩兩元素之間的絕對值差

# 定義max_value函式，用於問題2中Type3，計算使用者填答結果中出現最多次的結果
def max_value(row):
  lst = [i for i in list(row) if i != -1]         #排除值為-1(透過fillna將空值替換為-1)的值
  counter = Counter(lst)                      #將lst計數
  try:                                #如果取得最多計數key值失敗(即整row都是-1值，在函式第一步就已經去除)，則讓value == 空值
    value = max(counter)
  except:                   
    value = np.nan
  return counter.get(value)                     #回傳最大key值的出現次數

status_lst = []
question_63_fillna = question_63.fillna(-1) #透過fillna將空值替換為-1，方便計算(由於每個nan值皆是唯一值，無法用Counter計數)

for i in range(len(question_63_fillna)):
  counter = Counter(question_63_fillna.iloc[i])   #將每一row的填答計數
  platform_num = question_61['B2_Count'].iloc[i]   #取得每個受訪者有使用的平台總數
  gap_lst = gap(question_63_fillna.iloc[i, :])    #取得兩兩填答之間的絕對值間距
  max_result = max_value(question_63_fillna.iloc[i])  #計算每一row出現次數最多的填答結果
  # 選取的平台不超過2個且大於0個，標記為”PASS”
  if 0 < platform_num <= 2:
    print(f"{i} 狀態: PASS, 平台數: {platform_num}, 分數計數: {counter}, 原因: 平台數<=2, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("PASS")
  # 如果使用者在61題未填答或選擇第一選項則無法回答此題(即上題中的"B2_Count"欄位值等於Nan/0)，因此給予Nan值
  elif np.isnan(platform_num) or platform_num == 0:
    print(f"{i} 狀態: Nan, 平台數: {platform_num}, 分數計數: {counter}, 原因: 平台數==Nan/0, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append(np.nan)
  # 所有影音平台的喜好度都相同，透過counter得到此row總共有幾種填答結果，當填答結果種類==2(包含-1，也就是Nan值) 且 選取的平台超過2個 時，代表這些平台的填答分數都相同，因此給予Type1
  elif len(counter.keys()) == 2 and platform_num > 2:
    print(f"{i} 狀態: Type1, 平台數: {platform_num}, 分數計數: {counter}, 原因: 喜好度相同, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("Type1")
  # 當gap_lst取集合再以len()取長度的值 == 1 且 間隔不為0(以免變成Type1) 且 選取的平台超過2個 時， 即可得知填答間距相同， 因此給予Type2
  elif len(set(gap_lst)) == 1 and gap_lst[0] != 0 and platform_num > 2:
    print(f"{i} 狀態: Type2, 平台數: {platform_num}, 分數計數: {counter}, 原因: 填答間距相同, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("Type2")
  # 透過將max_result/platform_num來判斷是否超過9成喜好度相同(Q63總共只有9個欄位，8/9約等於0.88，因此以此資料即來說不可能出現Type3)
  elif max_result/platform_num >= 0.9 and platform_num > 2:
    print(f"{i} 狀態: Type3, 平台數: {platform_num}, 分數計數: {counter}, 原因: 超過9成喜好度填答相同, 填答資料: {list(question_63.iloc[i, :])}")  
    status_lst.append("Type3")
  else:
    print(f"{i} 狀態: PASS, 平台數: {platform_num}, 分數計數: {counter}, 原因: 不屬於任何規則, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("PASS")

question_63["Status"] = status_lst
question_63
```
