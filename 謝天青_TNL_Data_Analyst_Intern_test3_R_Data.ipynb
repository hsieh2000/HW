```python
import pandas as pd
import numpy as np
```

```python
RAW_DATA = pd.read_excel("/content/drive/MyDrive/關鍵評論網/TNL Data Analyst Intern_test3_R_Data.xlsx", sheet_name="Raw Data")
```

```python
question61_columns = [col for col, name in enumerate(RAW_DATA.columns) if "q0061" in name]
question_61 = RAW_DATA.iloc[:, question61_columns[0]:question61_columns[-1]+1]
question_61
```

```python
question_61["B2_Count"] = question_61.fillna(0).apply(lambda x : 0 if x[0] == 1 else (np.nan if sum(x) == 0 else sum(x[1:])), axis = 1)
```

```python
question63_columns = [col for col, name in enumerate(RAW_DATA.columns) if "q0063" in name]
question_63 = RAW_DATA.iloc[:, question63_columns[0]:question63_columns[-1]+1]
question_63
```

```python
from collections import Counter

def Sum_dict(Dict): 
  sum = 0
  for i in Dict.keys(): 
    if i == -1:
      pass
    else:
      sum = sum + Dict[i] 
  return sum

def gap(raw):
  lst = [i for i in list(raw) if i != -1]
  gap_lst = [abs(lst[i+1]-lst[i]) for i in range(len(lst)-1)]
  return gap_lst

status_lst = []
question_63_fillna = question_63.fillna(-1)

for i in range(len(question_63_fillna)):
  counter = Counter(question_63_fillna.iloc[i])
  platform_num = question_61['B2_Count'].iloc[i]
  gap_lst = gap(question_63_fillna.iloc[i, :])
  if platform_num <= 2:
    print(f"{i} 狀態: PASS, 平台數: {platform_num}, 分數計數: {counter}, 原因: 平台數<=2, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("PASS")
  elif np.isnan(platform_num):
    print(f"{i} 狀態: Nan, 平台數: {platform_num}, 分數計數: {counter}, 原因: 平台數==Nan, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("Nan")
  elif len(counter.keys()) == 2 and platform_num == counter.get(max(counter)) and platform_num > 2:
    print(f"{i} 狀態: Type1, 平台數: {platform_num}, 分數計數: {counter}, 原因: 喜好度相同, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("Type1")
  elif len(set(gap_lst)) == 1 and gap_lst[0] != 0 and platform_num > 2:
    print(f"{i} 狀態: Type2, 平台數: {platform_num}, 分數計數: {counter}, 原因: 填答間距相同, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("Type2")
  elif counter.get(max(counter))/Sum_dict(counter) >= 0.9 and platform_num > 2:
    print(f"{i} 狀態: Type3, 平台數: {platform_num}, 分數計數: {counter}, 原因: 超過9成喜好度填答相同, 填答資料: {list(question_63.iloc[i, :])}")  
    status_lst.append("Type3")
  else:
    print(f"{i} 狀態: PASS, 平台數: {platform_num}, 分數計數: {counter}, 原因: 不屬於任何規則, 填答資料: {list(question_63.iloc[i, :])}")
    status_lst.append("PASS")

question_63["Status"] = status_lst
question_63
```
