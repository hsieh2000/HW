# Part 1 合併資料

將 'Data_new' 與 'Data_rec' 合併成一份表，命名 'Data_all' 
* 合併時，請新增'**Ver**' 欄位來標記資料來源為'Data_new'或'Data_rec'
* 同個'**Email_enc**'有多筆填答時，請依 '**EndDate**' 保留最新一筆填答

```python
Data_new["Ver"] = "Data_new"
Data_rec["Ver"] = "Data_rec"
Data_all = pd.concat([Data_new,Data_rec])
Data_all = Data_all.sort_values(by = ["Email_enc","EndDate"], ascending = False)
Data_all = Data_all.drop_duplicates(subset=["Email_enc"], keep = "first")

print(f"Data_all資料筆數: {len(Data_all)}")
Data_all.head(10)
```

# Part 2 年齡分群
1. 在'Data_all' 新增'Age'欄位，存放填答者年齡
    * 請刪除年齡小於15歲的資料
2. 在'Data_all' 新增'P1'欄位，按照下表將年齡分組 
    | P1 | Age     |
    | ----------- | ----------- |
    | 1 | 15-24歲  |
    | 2 | 25-34歲  |
    | 3 | 35-44歲  |
    | 4 | 45歲以上 |

```python
bins = [15, 25, 35, 45, 101]
age_strata = ["15-24歲", "25-34歲", "35-44歲", "45歲以上"]
age_labels = {
    "15-24歲":1,
    "25-34歲":2,
    "35-44歲":3,
    "45歲以上":4
}  

do_age_strata = pd.cut(Data_all['q0002'], bins = bins, labels = age_strata, right = False)
```

```python
Data_all["Age"] = do_age_strata
print(f"原本資料筆數: {len(Data_all)}")
print(f'年齡小於15或是為空值的資料筆數: {len(Data_all[Data_all["q0002"]<15]) + len(Data_all[Data_all["q0002"].isna()])}')
Data_all = Data_all.dropna(subset = ["Age"])
Data_all["P1"] = Data_all["Age"].apply(lambda x: age_labels[x])
print(f"處理後資料筆數: {len(Data_all)}")

Data_all.head(10)
```

# Part 3 滿意度分組
q0063_0001 到 q0063_0009 為滿意度題，分別詢問受訪者對各個影音平台的滿意度。  
為方便分析，我們將0-10分重新分成三群，並將重新分群的結果，儲存在新的欄位。

1. 新增欄位：  
    在'Data_all'新增9個欄位，存放滿意度分組後的結果  
    欄位命名規則：'B5.{Question_Matrix_Option}_CODE'，  
    {Question_Matrix_Option} 請參照'QNR對照'  

2. 滿意度分群方式：
    |B5.{Question_Matrix_Option}_CODE|q0063_xxxx|
    |-|-|
    |1|0-6|
    |2|7-8|
    |3|9-10|

* 例1：如q0063_0001 的填答結果(滿意度)為5，則換算出的B5.meWatch_CODE為1
* 例2：如q0063_0003 的填答結果(滿意度)為8，則換算出的B5.Viki_CODE為2


```python
QNR_dict = {}
for i,j in zip(QNR_contrast['Question_ID'],QNR_contrast['Question_Matrix_Option']):
  QNR_dict[i] = j

QNR_dict.pop('q0002')
print(QNR_dict)

def re_cluster(rate):
  if 0 <= rate <= 6:
    return 1
  elif 7 <= rate <= 8:
    return 2
  elif 9 <= rate <= 10:
    return 3
  else:
    return rate

for i in QNR_dict.keys():
  lst = list(map(lambda x : re_cluster(x), Data_all[i]))
  Data_all[f"B5.{QNR_dict[i]}_CODE"] = lst

print(f"Data_all資料筆數: {len(Data_all)}")
Data_all.head(10)
```
