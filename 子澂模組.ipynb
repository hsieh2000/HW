```python
os.chdir('/content/drive/MyDrive/data/最新檔/')
print("Current working directory: {0}".format(os.getcwd()))
info = pd.read_csv("境外基金基本資訊.csv")
ex_rate = pd.read_csv("每日匯率20221001~20221031.csv")
Subscription_Redemption = pd.read_csv("申贖回資訊20221109.csv")
```

```python
info.columns
```

```python
info = info[['基金代碼', '基金名稱', '基金英文名稱', 'ISIN CODE', '總代理' ,'基金規模(百萬)','基金規模幣別']]
```

```python
currency = list(info['基金規模幣別'].unique())
```

```python
ex1031 = ex_rate[ex_rate["日期"]=='2022/10/31']
ex_dict = {}
for i in currency:
  for j in range(len(ex1031)):
    if i in ex1031["幣別"].iloc[j]:
      ex_dict.update({i:ex1031["賣出匯率"].iloc[j]})
```

```python
info["新台幣基金規模(百萬)"] = None
for i in range(len(info)):
  info["新台幣基金規模(百萬)"].iloc[i] = round(float(info["基金規模(百萬)"].iloc[i])*ex_dict[info["基金規模幣別"].iloc[i]],3)
```
